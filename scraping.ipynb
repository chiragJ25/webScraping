# Web Scraping using beautiful soup

### Get all the hyperlinks from Analytics Vidhya's homepage

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.analyticsvidhya.com/"
r = requests.get(url)
r
```

```python
doc = r.text
```

### Use html parser to parse the webpage and view it

```python

s = BeautifulSoup(doc,'html.parser')
#Prettify indents the parsed page in a 'pretty' and readable format
print(s.prettify())
```

### Find all the 'a' tags

```python
s.find_all('a')
```

### Get the hyperlinks within the anchor tag

```python
for hyperlink in s.find_all('a'):
    print(hyperlink.get('href'))
```
